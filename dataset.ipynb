```python
import pandas as pd
import gc
```

```python
def get_dataset(path:str):
    
    data=pd.read_csv(path)
    data_X=data.drop('label',axis=1)
    data_Y=data['label']
    
    del data
    gc.collect()
    
    return data_X,data_Y
```

```python
train_X, train_Y = get_dataset("archive/sign_mnist_train.csv")
test_X, test_Y = get_dataset("archive/sign_mnist_test.csv")
```
