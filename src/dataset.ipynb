Get image files from test data

```python
import pandas as pd
import matplotlib.pyplot as plt

def get_files(frame_X:pd.DataFrame, series_y:pd.Series):
    
    folder="../presentation_test_images/"
    
    for (index,image), label in zip(frame_X.iterrows(), series_y):
        plt.imsave(f'{folder}{response_transform[label]}({index}).png', image.to_numpy().reshape(28,28),cmap='gray')

get_files(test_X, test_Y)
```

Generate transform image

```python
from utility import sample_image, plot_image
import torchvision.transforms as T
from PIL import Image
import numpy as np
import pandas as pd

image,_ = sample_image(5)

image = Image.fromarray(image.to_numpy(dtype=np.float32).reshape((28,28)))

perspectiveImage = T.RandomPerspective(p=1)(image)
rotationImage = T.RandomRotation(degrees=45)(image)

plot_image(pd.Series(np.asarray(perspectiveImage).reshape(28*28)))
plot_image(pd.Series(np.asarray(rotationImage).reshape(28*28)))
```
