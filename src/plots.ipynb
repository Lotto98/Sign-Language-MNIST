```python
import matplotlib.pyplot as plt
import pandas as pd

from NeuralNetwork import NeuralNetwork
```

```python
data=pd.DataFrame(NeuralNetwork.load_stats("Classifier_3"))
data
```

```python
def plot_loss():
    
    fig,ax=plt.subplots(figsize=(20,10))
    
    #axs=[item for sublist in axs for item in sublist]
    
    ax.plot(data["epochs"],data["eval_losses"], color="blue")
    ax.plot(data["epochs"],data["train_losses"], color="orange")
    #ax.set_ylim(0,0.00001)
        
plot_loss()
```

```python
def plot_lr():
    
    fig,axs=plt.subplots(1,2, figsize=(15,15))
    
    #axs=[item for sublist in axs for item in sublist]
    
    for ax in axs:
        ax.plot(data["epochs"],data["current_lrs"])
        
plot_lr()
```

```python
def plot_accuracy():
    fig,axs=plt.subplots(1,2,figsize=(25,10))
    
    #axs=[item for sublist in axs for item in sublist]
    
    for ax in axs:
        ax.plot(data["epochs"],data["test_accuracies"])
        #ax.set_ylim(0.85,0.9)
        
plot_accuracy()
```
