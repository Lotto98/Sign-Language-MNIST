```python
def plot_loss():
    
    fig,ax=plt.subplots(figsize=(20,10))
    
    #axs=[item for sublist in axs for item in sublist]
    
    ax.plot(test["epochs"],test["eval_losses"], color="blue")
    ax.plot(test["epochs"],test["train_losses"], color="orange")
    #ax.set_ylim(0,0.00001)
        
plot_loss()
```

```python
def plot_accuracy():
    
    fig,ax=plt.subplots(figsize=(25,10))
    
    #axs=[item for sublist in axs for item in sublist]
    
    
    ax.plot(test["epochs"],test["test_accuracies"])
        #ax.set_ylim(0.85,0.9)
        
plot_accuracy()
```
